```python
def generate_colors(num):
    if num==1:
        colors=''
        lst = list('ABCDEF0123456789')
        for i in range(num):
            colors='#'+''.join(np.random.choice(lst, 6))
    else:
        colors = []
        lst = list('ABCDEF0123456789')

        for i in range(num):
            colors.append('#'+''.join(np.random.choice(lst, 6)))

    return colors
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Définition de la taille de la figure et de la résolution
plt.figure(figsize=(10, 5), dpi=200)

# Sous-graphique 1: Satisfaction des passagers par classe
plt.subplot(1, 2, 1)
plt.title('Satisfaction des Passagers - Classe', size=16, fontweight='bold')
sns.countplot(x='Class', hue='satisfaction', data=df, palette=['#055df5', '#c23a00'])
plt.ylabel('Nombre')  # Ajoutez des étiquettes si nécessaire
plt.xlabel('')
plt.yticks(np.arange(0, 70000, 10000))
plt.legend(fontsize='9', title='Satisfaction')

# Sous-graphique 2: Distance de vol par classe
plt.subplot(1, 2, 2)
plt.title('Classe - Distance de Vol', size=16, fontweight='bold')
sns.histplot(x='Flight Distance', hue='Class', data=df, kde=True)
plt.ylabel('')  # L'étiquette de l'axe y est omise pour le histplot car elle est généralement 'Count'
plt.xlabel('Distance')
plt.legend(['Économique Plus', 'Économique', 'Affaires'], fontsize='10', title='Classe')

# Ajustement de l'espacement entre les sous-graphiques
plt.subplots_adjust(wspace=0.4)

# Affichage de la figure
plt.show()
```



* Très peu de personnes voyagent en classe économique plus. Elles préfèrent
généralement l'Économique ou la Business.


* Plus de 80 % des passagers en classe économique sont soit Neutres, soit Insatisfaits. Cela nous montre que cette classe nécessite des améliorations.
*   Les passagers qui voyagent sur de longues distances préfèrent presque toujours la classe affaires, probablement parce qu'elle est plus confortable

```python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

plt.figure(figsize=(16,8))

plt.title('Satisfaction based on Delay',size=20,fontweight='bold')

sns.scatterplot(y='Departure Delay in Minutes',data=df,x='Arrival Delay in Minutes',hue='satisfaction',palette='husl')

plt.xlabel('Arrival Delay',size=18)
plt.ylabel('Departure Delay',size=18)

plt.legend(fontsize='13')
plt.show()
```

Plus le retard est long, plus les gens sont pas satisfaits (ou neutres), ce qui est attendu.


```python
df['satisfaction'] = df['satisfaction'].replace({'satisfied': 1, 'neutral or dissatisfied': 0})

# Calcul de la matrice de corrélation
corr_matrix = df.corr()

sns.set(style='darkgrid')
plt.figure(figsize=(20, 8))
sns.heatmap(corr_matrix, mask=np.triu(corr_matrix), cmap='BuPu', annot=True)
plt.show()
```

     

*   On constate qu'il existe une correspondance entre le retard à l'arrivée et le retard au départ, ce qui est logique et peut s'expliquer comme suit.
Si le vol des clients de la compagnie aérienne a été retardé d'un certaintemps au départ, alors le vol sera retardé d'à peu près le même temps à l'atterrissage
*   La nourriture et les boissons sont associées au divertissement en vol.


    
    

*   Il y a une corrélation entre le confort des sièges et la propreté.
   
*   La satisfaction des passagers dépend de toutes les caractéristiques du jeu de données (embarquement en ligne, confort des sièges, etc.) presque de manière égale.
