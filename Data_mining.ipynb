```python
def generate_colors(num):
    if num==1:
        colors=''
        lst = list('ABCDEF0123456789')
        for i in range(num):
            colors='#'+''.join(np.random.choice(lst, 6))
    else:
        colors = []
        lst = list('ABCDEF0123456789')

        for i in range(num):
            colors.append('#'+''.join(np.random.choice(lst, 6)))

    return colors
```

```python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

plt.figure(figsize=(16,8))

plt.title('Satisfaction based on Delay',size=20,fontweight='bold')

sns.scatterplot(y='Departure Delay in Minutes',data=df,x='Arrival Delay in Minutes',hue='satisfaction',palette='husl')

plt.xlabel('Arrival Delay',size=18)
plt.ylabel('Departure Delay',size=18)

plt.legend(fontsize='13')
plt.show()
```

Plus le retard est long, plus les gens sont pas satisfaits (ou neutres), ce qui est attendu.


```python
df['satisfaction'] = df['satisfaction'].replace({'satisfied': 1, 'neutral or dissatisfied': 0})

# Calcul de la matrice de corrélation
corr_matrix = df.corr()

sns.set(style='darkgrid')
plt.figure(figsize=(20, 8))
sns.heatmap(corr_matrix, mask=np.triu(corr_matrix), cmap='BuPu', annot=True)
plt.show()
```

     

*   On constate qu'il existe une correspondance entre le retard à l'arrivée et le retard au départ, ce qui est logique et peut s'expliquer comme suit.
Si le vol des clients de la compagnie aérienne a été retardé d'un certaintemps au départ, alors le vol sera retardé d'à peu près le même temps à l'atterrissage
*   La nourriture et les boissons sont associées au divertissement en vol.


    
    

*   Il y a une corrélation entre le confort des sièges et la propreté.
   
*   La satisfaction des passagers dépend de toutes les caractéristiques du jeu de données (embarquement en ligne, confort des sièges, etc.) presque de manière égale.
